```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("googleplaystore.csv") #chtenie

# очистка
df = df.dropna()

# топ категорий
top_categories = df['Category'].value_counts().head(10)
print(top_categories)

# средний рейтинг по категориям
rating_by_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print(rating_by_category.head(10))

# платные и бесплатные приложения
price_analysis = df.groupby('Type')['Rating'].mean()
print(price_analysis)
```

```python
top_categories.plot(kind='bar')
plt.title("Лучшие категории приложений")
plt.show()
```
